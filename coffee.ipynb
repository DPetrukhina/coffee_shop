## <a id='toc1_3_'></a>[Анализ продуктов, заказов и транзакций](#toc0_)
### <a id='toc1_3_1_'></a>[Количество транзакций по дням](#toc0_)

```python
pd.read_sql(
    """SELECT DATE_PART('YEAR', transaction_date::date) AS date_year,
              DATE_PART('MONTH', transaction_date::DATE) AS date_month,
              DATE_PART('DAY', transaction_date::DATE) AS date_day,
              COUNT(transaction_id)
       FROM coffe_shop.sales_reciepts
       GROUP BY transaction_date
       ORDER BY transaction_date;""", con=con_postgres)
```

### <a id='toc1_3_2_'></a>[Топ 10 самых продаваемых товаров каждый день](#toc0_)

```python
pd.read_sql(
    """SELECT pi2.transaction_date,
	   p.product_name,
	   SUM(quantity_sold) AS quantity_sold_per_day,
	   RANK() OVER(PARTITION BY transaction_date ORDER BY SUM(quantity_sold) DESC) AS rating
FROM coffe_shop.pastry_inventory pi2
JOIN coffe_shop.product p ON pi2.product_id=p.product_id 
GROUP BY p.product_name, pi2.transaction_date
ORDER BY pi2.transaction_date, SUM(quantity_sold) DESC;""", con=con_postgres)
```

### <a id='toc1_3_3_'></a>[Количество проданных и пропавших продуктов по неделям](#toc0_)

```python
pd.read_sql(
    """SELECT d.week_name,
	   SUM(pi2.quantity_sold) AS sold,
	   SUM(pi2.waste)  AS wasted
FROM coffe_shop.pastry_inventory pi2 
JOIN coffe_shop.dates d ON pi2.transaction_date::date=d.transaction_date
GROUP BY d.week_name
ORDER BY d.week_name;""", con=con_postgres)
```

### <a id='toc1_3_4_'></a>[Названия регионов , где продавался продукт “Columbian Medium Roast” с последней датой продажи в регионе](#toc0_)

```python
pd.read_sql(
    """SELECT neighborhood,
	          MAX(transaction_date) AS last_transaction
       FROM coffe_shop.sales_outlet so
       JOIN coffe_shop.sales_reciepts sr ON so.sales_outlet_id=sr.sales_outlet_id
       JOIN coffe_shop.product p ON sr.product_id=p.product_id
       WHERE p.product_name = 'Columbian Medium Roast'
       GROUP BY 1;""", con=con_postgres)
```

### <a id='toc1_3_5_'></a>[Продукты с упоминанием шоколада в названии](#toc0_)

```python
pd.read_sql(
    """SELECT *
       FROM coffe_shop.product 
       WHERE product_name LIKE '%%choco%%';""", con=con_postgres)
```

### <a id='toc1_3_6_'></a>[Сумма заказов в каждом городе и магазине](#toc0_)

```python
pd.read_sql(
    """SELECT so.store_city,
	   		  so.store_address,
	   		  round(SUM(sr.line_item_amount)) AS sale_amount
	   FROM coffe_shop.sales_outlet so
	   RIGHT JOIN coffe_shop.sales_reciepts sr ON so.sales_outlet_id=sr.sales_outlet_id
       GROUP BY so.store_address, so.store_city
	   ORDER BY sale_amount DESC;""", con=con_postgres)
```

### <a id='toc1_3_7_'></a>[Выручка и количество проданных товаров по магазинам за каждый день](#toc0_)

```python
pd.read_sql(
    """WITH 
    vitrin  AS 
	(
		SELECT
		transaction_date,
		sales_outlet_id,
		sum(line_item_amount) AS sum_line_item_amount,
		sum(quantity) AS sum_quantity
		FROM
		coffe_shop.sales_reciepts r
		GROUP BY
		transaction_date ,sales_outlet_id
	)
	SELECT 
	transaction_date,
	sales_outlet_id,
	sum_line_item_amount AS total_amount,
	sum_quantity AS total_quantity
	FROM 
	vitrin
 ORDER BY transaction_date;""", con=con_postgres)
```

## <a id='toc1_4_'></a>[Анализ клиентов](#toc0_)
### <a id='toc1_4_1_'></a>[Электронные адреса клиентов и их домены](#toc0_)

```python
pd.read_sql(
    """SELECT customer_name,
              email,
	          split_part(email, '@', 2) AS email_domen
       FROM coffe_shop.customer;""", con=con_postgres)
```

### <a id='toc1_4_2_'></a>[Справочник клиентов](#toc0_)

```python
pd.read_sql(
    """SELECT
	   split_part(customer_name, ' ', 1) as first_name,
	   split_part(customer_name, ' ', 2) as last_name,
       (CASE
	   		WHEN gender = 'M' THEN 'Male'
	   		ELSE 'Female'
	   END) AS gender,
       birthdate,
	   date_part('year',AGE(now(), birthdate::date)) AS customer_age,
       loyalty_card_number,
	   REPLACE(loyalty_card_number, '-', '') AS card_number_new,
	   COUNT(transaction_id) AS transaction_qty
FROM coffe_shop.customer c
LEFT JOIN coffe_shop.sales_reciepts sr 
ON c.customer_id = sr.customer_id
GROUP BY c.customer_name, loyalty_card_number, gender, birthdate
ORDER BY transaction_qty DESC, first_name;""", con=con_postgres)
```

### <a id='toc1_4_3_'></a>[Количество клиентов по поколениям](#toc0_)

```python
pd.read_sql(
    """SELECT generation,
	   COUNT(customer_id)
FROM coffe_shop.generations g
JOIN coffe_shop.customer c ON g.birth_year=c.birth_year
GROUP BY generation
ORDER BY count DESC;""", con=con_postgres)
```
